```python
# roots of quadratic equation
a=int(input("Enter x**2 coefficient"))
b=int(input("Enter x coefficient"))
c=int(input("Enter constant"))
from math import sqrt
det=(b*b-4*a*c) # determinant claculation
if det>=0:
  root_1=(-b+sqrt(det))/(2*a)
  root_2=(-b-sqrt(det))/(2*a)
  print(root_1, root_2)
else:
  print("Roots are complex.")
```

```python
# cell phone bill
m=int(input("Enter total number of talktime in minutes"))
t=int(input("Enter total number of text messages sent"))
base_price=15
cc=0.44
# m<=50 and t<=50
if m<=50 and t<=50:
  add_m_price=0
  add_t_price=0
elif m>50 and t>50:
  add_m_price=(m-50)*0.25
  add_t_price=(t-50)*0.15
elif m>50 and t<=50:
  add_m_price=(m-50)*0.25
  add_t_price=0
elif m<=50 and t>50:
  add_m_price=0
  add_t_price=(t-50)*0.15

price=base_price+add_m_price+add_t_price+cc
tax=price*(5/100)
bill=price+tax

print("Additional price on talktime",add_m_price,"$")
print("Additional price on text messages",add_t_price,"$")
print("tax",tax,"$.")
print("total bill payable",bill,"$.")
```

```python
# Next Day
# Assumption ----> we are not verifying leap year.
d=int(input("Enter day"))
m=int(input("Enter Month"))
y=int(input("Enter Year"))

g31=[1,3,5,7,8,10]
g30=[4,6,9,11]
if m==12 and d==31:
  d=1
  m=1
  y=y+1
elif (d==31 and m in g31) or (d==30 and m in g30) or (m==2 and d==28):
  d=1
  m=m+1
else:
  d=d+1
print("{}-{}-{}".format(d,m,y))
```
