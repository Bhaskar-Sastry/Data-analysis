# **3. Histogram:**

A histogram is a type of bar graph that represents the distribution of numerical data. It shows how often values (or ranges of values, called bins) occur in a dataset.

**Explanation:**

1. In a histogram, the x-axis represents intervals (bins) of data values (e.g., 0–10, 10–20, 20–30, etc.).

2. The y-axis represents the frequency — how many data points fall into each bin.

3. Unlike a regular bar chart (where each bar can represent different categories), a histogram is used only for continuous or numerical data.

**Example:**

Imagine you surveyed the ages of 100 people and got results like this:

* 0–10 years: 5 people

* 10–20 years: 12 people

* 20–30 years: 22 people

* 30–40 years: 30 people

* 40–50 years: 20 people

* 50–60 years: 11 people


A histogram of this data would have bars representing these age ranges, with the height of each bar showing how many people fall into each age group.

**Why Use Histograms?**

* To see the shape of your data (e.g., bell-shaped, skewed, uniform).

* To understand how data is distributed.

* To spot patterns, such as most common ranges or outliers.

```python
# Histogram using matplotlib
x=np.random.randint(1,100,size=100) # creating random data points
print(x)
plt.hist(x,bins=5)   # creating histogram
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

```python
# Histogram Using Seaborn
sns.histplot(x,bins=5)
plt.title("Histogram")
plt.show()
```

```python
# Histogram Using Plotly
fig=px.histogram(x=x,nbins=5)
fig.show()
```

# **Bar Plot:**

A bar plot is a chart that represents categorical data using rectangular bars. The height (or length) of each bar shows the value or frequency of the category it represents.

**Explanation:**

* In a bar plot, the x-axis typically shows categories (like colors, types of fruits, countries, etc.).

* The y-axis shows values such as counts, percentages, or any other measure related to those categories.

* The bars can be vertical or horizontal, depending on the orientation.

* The bars do not touch each other, because the categories are distinct and not continuous.

**Example:**

Let’s say you're counting how many pets people own:

* Dogs: 30

* Cats: 25

* Birds: 10

* Fish: 15

A bar plot would have 4 bars, one for each pet type, with heights matching their respective counts.

```python
# Bar plot using matplotlib
matches=[2,5,3,1,4]
venue=["Hyd","Mum","Che","Kol","Del"]
plt.bar(venue,matches)
plt.xlabel("Venue")
plt.ylabel("Matches")
plt.title("Bar Plot: Matches in each Venue ")
plt.show()
```

```python
# horizontal bar plot
plt.barh(venue,matches)
plt.xlabel("Venue")
plt.ylabel("Matches")
plt.title("Bar Plot: Matches in each Venue ")
plt.show()
```

```python
# barplot in seaborn
sns.barplot(x=venue,y=matches)
plt.xlabel("Venue")
plt.ylabel("Matches")
plt.title("Bar Plot: Matches in each Venue ")
plt.show()
```

```python
# horizontal bar plot in seaborn
sns.barplot(x=matches,y=venue)
plt.xlabel("Matches")
plt.ylabel("Venue")
plt.title("Bar Plot: Matches in each Venue ")
plt.show()
```

```python
# Bar plot in plotly
fig=px.bar(x=venue,y=matches)
fig.show()
```

# **Box Plot:**

A **box plot** is a graph that shows the **spread and distribution** of a dataset based on five summary statistics:  
**Minimum, First Quartile (Q1), Median (Q2), Third Quartile (Q3), and Maximum**.

**Explanation:**

A box plot helps you **visualize the range, center, and variability** of data.

**How it works:**

1. **Box**: Represents the **middle 50%** of the data (from Q1 to Q3).  
2.  **Line inside the box**: The **median** (middle value) of the dataset.
3.  **"Whiskers"**: Lines that extend from the box to the **minimum** and **maximum** values (excluding outliers).
4.  **Dots or stars outside the whiskers**: These are **outliers**, or unusually high or low values.

**Example:**

Imagine you have test scores out of 100 for a class:

- Minimum: 55  
- Q1 (25th percentile): 65  
- Median (50th percentile): 75  
- Q3 (75th percentile): 85  
- Maximum: 95

A box plot would:
- Draw a box from 65 to 85,
- Put a line at 75 (median),
- Extend whiskers to 55 and 95,
- Show any outliers if needed.

**Why Use a Box Plot?**

- To **see the spread** and **symmetry** of data.
- To quickly find the **median** and **range**.
- To **detect outliers** easily.
- To **compare distributions** across groups (e.g. test scores of two classes side-by-side).



**Summary:**


| Term        | Meaning                          |
|-------------|----------------------------------|
| Min         | Smallest value (not an outlier)  |
| Q1          | 25th percentile                  |
| Median (Q2) | Middle value (50th percentile)   |
| Q3          | 75th percentile                  |
| Max         | Largest value (not an outlier)   |


```python
# Box plot using matplotlib
print(x)
plt.boxplot(x)
plt.show()
```

```python
# horizontal box plot using matplotlib
plt.boxplot(x,vert=False)
plt.show()
```

```python
# Box plot using seaborn
sns.boxplot(x)
plt.show()
```

```python
# Box plot using plotly
fig=px.box(x)
fig.show()
```

# **Pie Chart:**

A **pie chart** is a circular graph used to represent **parts of a whole**. Each "slice" of the pie shows a **category’s proportion** or **percentage** of the total.

**Explanation:**

- The entire pie represents **100%** of your data.
- Each **slice** corresponds to a **category** and its size shows **how much it contributes** to the total.
- The **angle or area** of each slice is proportional to the value it represents.

**Example:**

Let’s say you surveyed favorite fruits among 100 people:

- Apples: 40  
- Bananas: 30  
- Oranges: 20  
- Grapes: 10  

A pie chart would:

- Show 4 slices,
- Apples take up 40% of the pie,
- Bananas 30%, and so on.

**Why Use a Pie Chart?**

- To show **percentages** or **proportions** clearly.
- To highlight **which category is the biggest** (or smallest).
- To provide a **quick visual impression** of how a whole is divided.

**When *Not* to Use a Pie Chart:**

- When you have **too many categories** — it gets cluttered.
- When values are **too close together** — it's hard to compare.
- In those cases, a **bar chart** might be a better choice.

**Summary:**


| Feature           | Pie Chart                    |
|-------------------|------------------------------|
| Data Type         | Categorical                  |
| Represents        | Part-to-whole relationships  |
| Total Value       | 100%                         |
| Good for          | Showing proportions clearly  |


```python
# pie chart using matplotlib
venue=["Hyd","Mum","Che","Kol","Del"]
matches=[20,15,19,18,24]
plt.pie(matches,labels=venue,autopct="%0.2f%%")
plt.show()
```

```python
# Pie chart using seaborn
sns.set(style="darkgrid")
plt.pie(matches,labels=venue,autopct="%0.2f%%")
plt.show()
```

```python
# Pie Chart using Plotly
fig=px.pie(values=matches,names=venue)
fig.show()
```

# **KDE Plot:**

A **KDE plot** (Kernel Density Estimate plot) is a **smooth curve** that shows the **probability density** of a continuous variable — basically, it estimates the **distribution** of your data.

**Explanation:**

- Think of a KDE plot as a **smoothed version of a histogram**.
- It shows where data points are **concentrated** (peaks) and where they are **sparse** (valleys).
- The area under the curve adds up to **1**, because it represents a **probability distribution**.

- Instead of using bars (like in a histogram), the KDE plot uses a **kernel function** (usually a Gaussian curve) to estimate how likely values are to occur in different ranges.

**Example:**

Imagine you’re plotting the heights of 100 students:

- A KDE plot would show a smooth curve where:
  - The **tallest peak** might be around 170 cm (the most common height).
  - It gradually rises and falls, giving you a sense of how spread out or concentrated the heights are.

**Why Use a KDE Plot?**

- To get a **smooth, clean view** of the data distribution.
- To **compare multiple distributions** (you can plot several KDEs on the same graph).
- To **understand the shape** of your data — like whether it's **normal**, **skewed**, or **multi-modal** (has more than one peak).

**KDE vs Histogram:**

| Feature         | Histogram                        | KDE Plot                          |
|-----------------|----------------------------------|-----------------------------------|
| Appearance      | Bars                             | Smooth curve                      |
| Interpretation  | Count/frequency                  | Probability density               |
| Smoothness      | Choppy (depends on bin size)     | Smooth (depends on bandwidth)     |
| Customizable?   | Bins                             | Kernel & bandwidth                |

**Bandwidth**

The **bandwidth** controls how smooth the KDE is:
- Small bandwidth = curve is more wiggly (more detail).
- Large bandwidth = smoother curve (less detail).

```python
# KDE plot using seaborn
print(x)
sns.kdeplot(x)
plt.show()
```

# **Boxen Plot:**

A **boxen plot** is an advanced version of a box plot, designed to **visualize large datasets** and better show the **distribution and tails (extremes)** of the data.

**Explanation:**

Like a box plot, a boxen plot shows:

- The **median** (middle value),
- The **quartiles** (25%, 50%, 75%),
- The **spread** of the data.

- But the **boxen plot goes further** by showing **additional layers of data beyond the quartiles**, especially in the **tails** — where the data is less dense.

- It divides the data into **"letter values"** (Q1, Q2, Q3, Q4, Q5, etc.) — hence the nickname *letter-value plot*. Each level of the plot provides **more detail about the distribution**, especially useful when you have **a lot of data**.

**Example:**

Imagine you have salary data for 10,000 people:

- A normal box plot would only show the median and a few quartiles.
- A **boxen plot** would show how those high earners and outliers are spread out in more detail — not just with a single whisker.

**Why Use a Boxen Plot?**

- Best for **very large datasets**.
- Reveals **fine details in the tails** of the distribution.
- Helps spot **skewness** or unusual patterns in high/low values.

**Comparison: Box Plot vs Boxen Plot**

| Feature         | Box Plot                        | Boxen Plot                          |
|-----------------|----------------------------------|-------------------------------------|
| Best for        | Small to medium datasets         | Large datasets                      |
| Tails shown?    | Limited                          | Detailed                            |
| Number of boxes | Usually 1 central box            | Multiple nested boxes               |
| Focus           | Median and quartiles             | Full shape of the distribution      |

```python
# Boxen Plot using seaborn
sns.boxenplot(x)
```

# **Violin Plot:**

A **violin plot** is a data visualization that combines the features of a **box plot** and a **KDE plot** (Kernel Density Estimate). It shows the **distribution, probability density, and summary statistics** of a dataset.

**Explanation:**

- A violin plot looks like a **mirrored KDE plot** with a **box plot inside it**:

- The **width** of the "violin" at any point shows the **density** (how frequent values are at that level).
- The **center** includes a **box plot** showing the **median** and **quartiles**.
- The shape of the violin helps you understand whether the data is **symmetric**, **skewed**, or **multi-modal** (has multiple peaks).

**Example:**

Let’s say you're plotting test scores of students:

- If most students score around 80, the violin will be **widest near 80**.
- If some score very low or very high, the **tails** of the violin will extend and show those extremes.
- Inside the violin, a box shows the **median and quartile range**, just like a box plot.

**Why Use a Violin Plot?**

- To **see both summary stats and the full distribution** at the same time.
- Better than a box plot when you want to **visualize the shape** of the data.
- Great for **comparing multiple groups** side by side.

**Comparison: Violin vs Box vs KDE**

| Feature         | Box Plot             | KDE Plot               | Violin Plot                       |
|-----------------|----------------------|------------------------|------------------------------------|
| Median/Qs       | ✅ Yes               | ❌ No                  | ✅ Yes (in the center box)         |
| Density Shape   | ❌ No                | ✅ Yes                 | ✅ Yes                             |
| Symmetry Shown  | ❌ No                | ✅ Yes                 | ✅ Yes                             |
| Good for Groups | ✅ Yes               | ✅ Sometimes           | ✅✅ Great for multiple groups      |

# **Count Plot:**

- A **count plot** is a type of bar plot that shows the **number of occurrences (count)** of each category in a categorical variable.

- It is one of the simplest and most useful plots when you want to **see how many times each category appears** in your dataset.

**Explanation:**

- The **x-axis** (or y-axis) shows the **categories** (e.g., gender, colors, animal types).
- The **height** of the bar shows **how many data points** fall into each category.
- It’s essentially the **visual form of `value_counts()`** in pandas.

- Count plots are often used with **Seaborn** using the function `sns.countplot()`.

**Example:**

Imagine you have a column for "Pet Type" with values like:

`["Dog", "Cat", "Dog", "Bird", "Cat", "Dog"]`

A count plot would show:
- Dog: 3  
- Cat: 2  
- Bird: 1

And plot 3 bars with heights matching those counts.

**Why Use a Count Plot?**

- To **quickly understand how data is distributed** across categories.
- Super useful for **categorical variables**.
- Helps detect **imbalanced classes** (like in classification problems).
- Very handy in **exploratory data analysis (EDA)**.

**Count Plot vs Bar Plot:**

| Feature           | Count Plot                             | Bar Plot                          |
|-------------------|-----------------------------------------|------------------------------------|
| Input             | Raw categorical data                   | Pre-aggregated data (values + labels) |
| Used for          | Frequency of categories                | Any numerical values by category   |
| Example Tool      | `sns.countplot()`                      | `plt.bar()`, `sns.barplot()`       |


**Summary:**

- **Count plot = bar chart of category counts**
- Fast, easy, and great for getting a feel for your data.
