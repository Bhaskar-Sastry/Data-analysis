**Understanding Missing Values in Student Performance Data**  

Missing values in datasets often indicate gaps in data collection, non-response, or external factors preventing measurement. In the **National Achievement Survey 2021**, missing values in student performance metrics (Maths %, Reading %, Science %, and Social %) suggest that **performance records were not available**, likely because students were absent during the assessment or their responses were not documented.  

Since the core objective of academic analysis is to understand student **learning outcomes**, missing performance data undermines the ability to assess correlations with other variables such as **internet usage, parental education, household responsibilities, or library access**. Without recorded performance metrics, it is impossible to make meaningful conclusions about the relationship between **student behavior, socioeconomic factors, and academic success**.  
  
Given the nature of the missing values, analyzing additional student attributes **without performance records would be ineffective**. The best approach is to **first handle missing data**, either by filtering out absent students or using data imputation strategies to reconstruct partial records. Only then can a structured and valid analysis take place.

#### **11. How does gender affect student performance?**


```python
# Select only performance columns and gender
performance_columns = ['Maths %', 'Reading %', 'Science %', 'Social %']
gender_performance = nas.groupby('Gender')[performance_columns].mean()

# Display results
print("Average student performance based on gender:")
print(gender_performance)
```

```python
# Visualizing gender-based performance differences
plt.figure(figsize=(10, 6))
gender_performance.T.plot(kind='bar', figsize=(10, 6))

# Customize plot
plt.xlabel("Subjects", fontsize=14, fontweight='bold')
plt.ylabel("Average Performance (%)", fontsize=14, fontweight='bold')
plt.title("Comparison of Student Performance by Gender", fontsize=16, fontweight='bold')
plt.legend(title="Gender")

# Show the plot
plt.xticks(rotation=0)  # Keep labels horizontal for readability
plt.show()
```

- **Gender does not significantly affect academic performance** in most subjects.  
- **Girls excel slightly in reading & science, while boys show an edge in social studies.**  
- These findings challenge gender-based learning stereotypes and emphasize **equal academic potential across genders**.

#### **15. How do student scores in different subjects relate to each other?**

- Students excelling in **one subject tend to do fairly well in others**, but some relationships are stronger than others.  
- The **highest connection is between Maths & Social Studies**, which is an interesting insight because **logical reasoning may support structured thinking in history/civics**.  
- The **lowest relationship is between Maths & Reading (0.3143)**, suggesting numerical and language skills are **less dependent** on each other.

#### **17. How does student absenteeism (missing scores) impact their engagement in learning activities?**  


```python
nas.columns
```

```python
# Create a new column 'Performance Status' based on absenteeism in subjects
nas['Performance Status'] = nas[['Maths %', 'Science %', 'Reading %', 'Social %']].isna().sum(axis=1).apply(lambda x: 'Absent' if x > 2 else 'Present')

# Group data by 'Performance Status' and count engagement factors
engagement_counts = nas.groupby('Performance Status')[['Use Internet', 'Library use', 'Help in household']].count()

# Display results
print("Comparison of student engagement based on performance status:")
print(engagement_counts)
```

```python
# Create a bar plot
plt.figure(figsize=(10, 6))
engagement_counts.plot(kind='bar', figsize=(10, 6), colormap="Set2")

# Customize plot
plt.xlabel("Performance Status", fontsize=14, fontweight='bold')
plt.ylabel("Number of Students Engaged", fontsize=14, fontweight='bold')
plt.title("Impact of Student Absenteeism on Learning Engagement", fontsize=16, fontweight='bold')
plt.legend(title="Engagement Factors")

# Show the plot
plt.xticks(rotation=0)
plt.show()
```

- **Students who are present in assessments show higher engagement across all learning resources.**  
- **Absent students may face challenges that limit their participation in educational activities.**  
- **Lower engagement among absent students could indicate potential learning barriers that need attention.**
