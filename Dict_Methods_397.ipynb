# **Dictionary**

- It stores the data using key - value pairs.

- syntax:  **variable_name = {key1 : value1, key2 : value2, key3 : value3,.....}**

- Dictionary is called a **mapping** data type.

- Dictionary is a **Mutable** Data Type.

- **Key - value pair** together is called an element in a dictionary.

```python
# Let us create dictionary
x={"python":96,"eda":92,"stats":87}
print(x)
print(type(x))
```

```python
y={96:"python",92:"eda",87:"stats"}
print(y)
print(type(y))
```

```python
z={"c1":"Ravi","c2":"rahul","c3":"rohith"}
print(z)
print(type(z))
```

```python
runs={"virat":[34,49,56,17,26],"rohith":[45,46,49,21,17],"msd":[23,11,14,45,33]}
print(runs)
print(type(runs))
```

# **Dictionary Methods:**

1.   **len():** Number of elements in the dictionary
2.   **clear():** it removes the data
3.   **del():** it deletes data
4.   **get():** to access the data from dictionary
5.   **dict():** It converts the data into a dictionary
6.   **keys():** To get all the keys present in the dictionary
7.   **values():** To get all the values present in the dictionary
8.   **items():** to get the values from the dictionary.

```python
# len()
d={"first":"a","sec":"b","third":"c"}
print(d)
print(type(d))
```

```python
print(len(d))
```

```python
x={"first":"a","sec":"b","third":"c"}
print(x)
```

```python
# clear()
x.clear()
```

```python
print(x)
```

```python
# del()
del x
```

```python
print(x)
```

```python
# get()
x=[10,20,30]
print(x)
print(type(x))
```

```python
# get
d={'first': 'a', 'sec': 'b', 'third': 'c'}
print(d)
```

```python
print(d.get("first"))
```

```python
print(d.get("third"))
```

```python
# dict(): if data is given as list of tuples, it converts into dict.
data=[(1,"a"),(2,"b"),(3,"c"),(4,"d")]
d=dict(data)
print(d)
```

```python
# items(): it returns list of tuples from a dictionary
d.items()
```

```python
# keys(): it return all the keys from dictionary
print(d)
print(d.keys())
```

```python
# values(): it return all the values from disctionary
print(d)
print(d.values())
```

```python
# values()
print(d)
print(d.values())
```
