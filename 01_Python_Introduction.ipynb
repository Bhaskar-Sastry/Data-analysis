# **Introduction to Python Programming**

## **Programming Language:**

- A **programming language** is a set of rules and syntax that allows humans to write **instructions for computers**.

- Programming Language acts as a **translator:** humans write **code** in a programming language, and the computer interprets it to **perform tasks**.

##**Program Execution Process:**

The **program execution process** refers to how a computer runs a program written in a programming language.

image.png

**1. Writing the Program**
   - A programmer writes code in a high-level language (like Python, C, or Java).
   - The source code is saved in a file (e.g., program.py, program.c).

**2. Translation to Machine Code**
   - Computers only understand **machine language** (binary instructions), so the code must be translated.
   - **Compilation** (for compiled languages like C, C++):
     - The compiler converts source code into an **executable** machine code file.
     - Errors are detected during compilation and must be fixed.
   - **Interpretation** (for interpreted languages like Python, JavaScript):
     - The interpreter translates and executes the code **line by line**.
     - Slower than compilation but allows flexibility.

**3. Loading into Memory**
   - The **Operating System (OS)** loads the compiled or interpreted code into memory.
   - The CPU retrieves instructions and begins execution.

**4. Execution by the CPU**
   - The CPU fetches instructions, decodes them, and executes them.
   - The program interacts with system resources (memory, disk, network).

**5. Handling Inputs & Outputs**
   - The program may take input from the user, process it, and produce output.
   - Output can be displayed on a screen, saved to a file, or sent over a network.

**6. Termination or Continuous Execution**
   - The program finishes execution after completing its task.
   - Some programs run indefinitely (e.g., web servers) until manually stopped.

##**Types of Programming Languages:**

- Programming languages can be classified based on different criteria, such as  
       
       (a) Abstraction Level
       (b) Programming Paradigm
       (c) Execution Method
       (d) Usage.

image.png

Programming languages can be classified based on different criteria, such as abstraction level, paradigm, execution method, and usage.

### **1. Based on Abstraction Level**
   - **Low-Level Languages**: Closer to hardware, requiring deep knowledge of computer architecture.
     - **Machine Language** (Binary code  directly executed by CPU).
     - **Assembly Language** (Symbolic representation of machine instructions).
   - **High-Level Languages**: More human-readable and abstracted from hardware.
     - Examples: Python, Java, C++

### **2. Based on Programming Paradigm**
   - **Imperative Languages**: Execute commands in a sequential manner.
     - Examples: C, Java, Python
   - **Declarative Languages**: Describe _what_ should be done rather than _how_.
     - Examples: SQL, Prolog
   - **Procedural Languages**: Based on structured programming with procedures/functions.
     - Examples: C, Pascal
   - **Object-Oriented Languages**: Organize code into objects for modularity and reuse.
     - Examples: Java, C++, Python
   - **Functional Languages**: Focus on mathematical functions and immutability.
     - Examples: Lisp, Haskell, Erlang
   - **Logic Programming Languages**: Use facts and rules for computation.
     - Examples: Prolog, Datalog

### **3. Based on Execution Method**
   - **Compiled Languages**: Code is converted into machine code before execution.
     - Examples: C, C++, Rust
   - **Interpreted Languages**: Code is executed line-by-line by an interpreter.
     - Examples: Python, JavaScript, PHP
   - **Hybrid Languages**: Use both compilation and interpretation.
     - Examples: Java (compiled into bytecode, then interpreted by JVM)

### **4. Based on Usage**
   - **General-Purpose Languages**: Used for a wide range of applications.
     - Examples: Python, Java, C++
   - **Scripting Languages**: Used for automation, web development, etc.
     - Examples: JavaScript, Perl, Ruby
   - **Markup Languages**: Define document structure and layout.
     - Examples: HTML, XML
   - **Query Languages**: Used for database queries.
     - Example: SQL
   - **Domain-Specific Languages (DSLs)**: Designed for specialized tasks.
     - Examples: MATLAB (scientific computing), Verilog (hardware description)

#**Python Programming Language:**

#**Brief History of Python:**

- Python was created by **Guido van Rossum** in the late 1980s and officially released in **1991**.

- It was designed to be **simple, readable, and versatile**, allowing developers to write clean code with **fewer lines**.

### **Key Milestones in Python's History**

- **1989**  Guido van Rossum begins developing Python at the Centrum Wiskunde & Informatica (CWI) in the Netherlands.

- **1991**  Python **1.0** is released, featuring exception handling and modules.

- **2000**  Python **2.0** introduces list comprehensions and garbage collection.

- **2008**  Python **3.0** arrives, bringing significant improvements but breaking backward compatibility.

- **2018**  Python **2.7 support ends** to encourage full migration to Python 3.

- **Present**  Python continues to evolve **(3.13)**, widely used in **AI, machine learning, web development, and automation**.

#**Python Features:**

image.png

**Simplicity & Readability:**

- Python's syntax is straightforward and resembles English, making it easier for beginners to learn and use.

- The language allows developers to write programs with fewer lines of code, reducing complexity.

**Interpreted Language:**

- Python is an interpreted language, meaning its code is executed line by line rather than being compiled beforehand.

- This makes debugging and testing more convenient.

**Dynamically Typed:**

- Python does not require explicit type declaration for variables.

- Instead, the type is determined at runtime, making it flexible and reducing the need for manual type management.

**High-Level Language:**

- Being a high-level language, Python abstracts complex low-level operations such as memory management and system interactions.

- This allow programmers to focus on logic rather than system specifics.

**Object-Oriented & Procedural:**

- Python supports both object-oriented programming (OOP) and procedural programming.

- OOP allows data and behavior to be organized into reusable objects, while procedural programming follows a structured approach.

**Extensive Libraries & Modules:**

- Python has a vast collection of built-in and third-party libraries that simplify development across various domains.

- Some domains including web development, data science, artificial intelligence, cybersecurity, and networking.

**Multi-Platform Support:**

- Python is cross-platform, meaning it can run seamlessly on different operating systems.

- Which Includes Windows, Mac, and Linux without requiring modifications.

**Memory Management & Garbage Collection:**

- Python has an automatic memory management system that includes garbage collection.

- This helps remove unused objects, optimizing performance without requiring manual intervention.

**Strong Community Support:**

- Python has one of the largest programming communities worldwide.

- This active support network ensures easy access to solutions, tutorials, and updates, making learning and troubleshooting more efficient.

**Integration & Extensibility:**

- Python integrates well with other programming languages such as C, C++, and Java, making it ideal for projects that require multi-language interoperability.

- It also supports third-party extensions for additional functionality.
