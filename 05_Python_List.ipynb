# **Python List**

- List is used to store multiple elements in a single variable.

- syntax: variable_name = [e1,e2,e3,..........]

- total number of elements in a list is called **length (size)** of a list.

- **len()** is used to get the length of any list.

- List allows **multiple data type** as its elements.

- two or more lists are concatenated using **+** operator.

- **multiplication operator** is used to repeate the elements in a list.

- list is a **sequential Data Type**. Hence, **indexing** can be used.

- **index:** a numerical value assigned to every element in a list.

- If **positive numbers** are used as index values, it is called **positive indexing**.

- If **negative numbers** are used as index values, it is called **negative indexing.**

- **slicing:** Accessing elements from the list.

- **Slicing** is two types: (a) forward direction slicing (b) reverse direction slicing.

- List without any elements is called an **empty (null)** list.
