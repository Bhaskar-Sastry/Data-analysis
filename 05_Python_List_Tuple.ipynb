# **Python List**

- List is used to store multiple elements in a single variable.

- syntax: variable_name = [e1,e2,e3,..........]

- total number of elements in a list is called **length (size)** of a list.

- **len()** is used to get the length of any list.

- List allows **multiple data type** as its elements.

- two or more lists are concatenated using **+** operator.

- **multiplication operator** is used to repeate the elements in a list.

- list is a **sequential Data Type**. Hence, **indexing** can be used.

- **index:** a numerical value assigned to every element in a list.

- If **positive numbers** are used as index values, it is called **positive indexing**.

- If **negative numbers** are used as index values, it is called **negative indexing.**

- **slicing:** Accessing elements from the list.

- **Slicing** is two types: (a) forward direction slicing (b) reverse direction slicing.

- List without any elements is called an **empty (null)** list.

- **List is a Mutable Data Type** (add new element, delete element, modify)

```python
# let us create a list
x=[10,20,30,40,50]
print(x)
print(type(x))
print(len(x))
```

# **List Methods (Functions/Builtin Methods):**



1.   **len():** it gives count of elements in a list.

2.   **count():** it gives number of occurances of an element in a list.

3.   **append():** it adds new element at the end of list.

4.   **insert():** adds element at required position.

5.   **extend():** It is used to merge the data in two lists.

6.   **remove():** It removes an element from list.

7.   **pop():** It remove an element based on the index.

8.   **reverse():** It reverses the order of all the elements in a list.

9.   **sort():** it sorts the elements.

10.  **sorted():** it sorts the list by preserving original order.

11.  **min():** minimum value in all the elements of a list.

12.  **max():** maximum values in all the elements of a list.

13.  **sum():** sum of all the elements in a list.

14.  **list():** it converts a sequence (ex:string) into a list.

# **Tuple Data Type**

- Tuple is used to store multiple elements in a single variable.

- syntax: variable_name = (e1,e2,e3,..........)

- total number of elements in a tuple is called **length (size)** of a list.

- **len()** is used to get the length of any tuple.

- tuple allows **multiple data type** as its elements.

- two or more tuples are concatenated using **+** operator.

- **multiplication operator** is used to repeate the elements in a tuple.

- tuple is a **sequential Data Type**. Hence, **indexing** can be used.

- **index:** a numerical value assigned to every element in a tuple.

- If **positive numbers** are used as index values, it is called **positive indexing**.

- If **negative numbers** are used as index values, it is called **negative indexing.**

- **slicing:** Accessing elements from the tuple.

- **Slicing** is two types: (a) forward direction slicing (b) reverse direction slicing.

- tuple without any elements is called an **empty (null)** list.

- **Tuple is an Immutable Data Type** (cannot modify a tuple)

```python
x=(10,20,30,40)
print(x)
print(type(x))
print(len(x))
```

```python
y=(12,"abc",99.88)
print(y)
print(type(y))
print(len(y))
```
