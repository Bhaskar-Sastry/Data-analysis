# **2. Assignment Operators**

- Assignment operators allow values to be assigned to variables in programming languages.

- The most basic assignment operator is **=** which directly assigns a value.

- For example, if you set a **variable equal to a number**, that number is stored within the variable.

#### **Compound Assignment Operators**

- These operators **combine arithmetic** or bitwise operations **with assignment**, making code more concise.

##### **Arithmetic-based Assignment Operators**

- **Addition assignment (+=):** increases a variable's value.

- **Subtraction assignment (-=):** decreases a variable's value.

- **Multiplication assignment (*=):** scales a variable by a factor.

- **Division assignment (/=):** reduces a variable through division.

- **Modulus assignment (%=):** retains only the remainder of a division.

- **Exponentiation assignment (**=):** raises a number to a power (specific to languages like Python).


### **Benefits of Using Assignment Operators**

- **Improved Readability:** Makes statements concise and easy to understand.

- **Optimized Performance:** Reduces redundant calculations.

- **Simplified Code Management:** Allows easy modifications of complex expressions.

# **4. Logical (Boolean) Operators:**

- Boolean operators are essential in programming and logic-based systems, allowing the evaluation of expressions that return **True** or **False**.

- These operators are primarily used in conditional statements, loops, and decision-making processes.

#### **Types of Boolean Operators**

- Boolean operators typically work with truth values and help in logical comparisons.

##### **and Operator:**

- Returns **True** only if both conditions are **True**.

- If at least one condition is **False**, the result is **False**.

##### **or Operator:**

- Returns **True** if at least one condition is **True**.

- Only returns **False** if both conditions are **False**.

##### **not Operator:**

- Reverses the truth value.

- If a condition is **True**, applying **not** makes it **False** and vice versa.

##### **Exclusive OR (xor):**

- Returns **True** only if **one** of the conditions is True, but **not both**.

### **Why Use Boolean Operators?**

- **Efficient Decision-Making:** Controls execution flow in programs.

- **Data Filtering:** Helps refine results in queries and searches.

- **Logical Computation:** Essential in AI, algorithms, and automation.

# **5. Identity Operators:**

- Identity operators in Python are used to compare whether two variables **refer to the same memory location**.

- Unlike comparison operators (==), which check for equality in values, identity operators determine if two objects are **identical** (i.e., they occupy the same memory space).

### **Types of Identity Operators**

Python provides two identity operators:

1. **is**  Returns **True** if two variables reference the same object.

2. **is not**  Returns **True** if two variables reference different objects.

### **How Identity Operators Work**

- Identity operators are particularly useful in checking object references rather than value equality.

#### **Difference Between == and is**

- The **==** operator checks whether two variables **have the same value**.

- The **is** operator checks whether two variables **point to the same memory address**.


### **Usage Across Data Types**

- **Usage With integers and strings:**

- Python optimizes memory allocation, so small integer values and short strings may return **True** when checked with **is**.

- **Usage With lists and dictionaries:**

- Even if two lists contain the same elements, they are distinct objects unless explicitly assigned the same reference.


### **Why Use Identity Operators?**

- **Efficient Memory Checks:** Helps determine if variables share the same object reference.

- **Optimizes Object Comparisons:** Useful for checking singleton objects like **None**.

- **Ensures Correct Object Handling:** Prevents unintended modifications by verifying object uniqueness.

# **6. Membership Operators:**

- Membership operators in Python are used to check whether a value or object is present within a sequence like a list, tuple, set, dictionary, or string.

- These operators return a **Boolean result (True or False)**.

#### **Types of Membership Operators**

Python provides two membership operators:

1. **in**  Checks if a value exists in a sequence.

2. **not in**  Checks if a value does **not** exist in a sequence.


### **How Membership Operators Work**

Membership operators evaluate sequences like lists, tuples, sets, strings, and dictionaries.

#### **Usage with Lists and Tuples**

- The **in** operator verifies whether an item exists in a list or tuple.

- The **not in** operator confirms that an item is absent.

#### **Usage with Strings**

- The **in** operator searches for a substring within a string.

- The **not in** operator ensures a substring is missing.

#### **Usage with Dictionaries**

- The **in** operator checks if a **key** exists in a dictionary.

- The **not in** operator checks if a key is missing (it does not check values directly).

---

### **Why Use Membership Operators?**
- **Efficient Searching:** Quickly determines if an element is present.

- **Simplifies Code:** Eliminates the need for loops in many cases.

- **Flexible:** Works with various data structures, including lists, tuples, and dictionaries.

# **7. Bit Wise Operators:**

- Bitwise operators allow you to manipulate individual bits in binary numbers rather than working with the whole number.

- Computers store numbers in binary (0s and 1s), and bitwise operations help in performing efficient calculations and data processing.

### **Types of Bitwise Operators**

1. **Bitwise AND (&):**

   - Compares two numbers bit by bit.

   - Only gives **1** where both corresponding bits are **1**; otherwise, it results in **0**.

   - Used in bit masking to check if a specific bit is set.

2. **Bitwise OR (|):**  

   - Compares two numbers bit by bit.

   - Gives **1** if at least one of the corresponding bits is **1**.

   - Useful in setting specific bits.

3. **Bitwise XOR (^):**

   - Compares two numbers bit by bit.

   - Gives **1** where the corresponding bits are different.

   - Used in toggling bits and error detection techniques.

4. **Bitwise NOT (~):**  

   - Inverts the bits of a number.

   - Turns **0s** into **1s** and **1s** into **0s**.

   - Used in operations requiring a binary inversion.

5. **Bitwise Left Shift (<<):**

   - Shifts bits of a number to the left.

   - Each shift essentially multiplies the number by 2.


6. **Bitwise Right Shift (>>):**

   - Shifts bits of a number to the right.

   - Each shift divides the number by 2.


### **Why Are Bitwise Operators Important?**

- They make computations faster compared to traditional arithmetic operations.

- Used in **image processing, cryptography, networking protocols, and embedded systems**.

# **IF Statement:**

- The if statement expresses selective execution in Python.

- It is the first control statement, and simple as it is, it is quite powerful.

- Selective execution is one of the primary control mechanisms in general programming.

- It is a critical part of enabling computers to perform
complex tasks.



if boolean expression:

< > if suite

The **basic if statement** allows you to do the following:

1. Evaluate boolean expression, yielding either **True or False**.

2. If boolean expression yields **True**,

  (a) Execute the Python suite of code indented under the if, the if suite.

  (b) Once the indented code is executed, continue with any Python code after the indented code (that is, any code following the if at the same indentation as the if itself ).
3. If boolean expression yields **False**,

  (a) Ignore any code indented under the if, (that is, do not execute it).

  (b) Continue with any Python code after the indented code (that is, any code following the if at the same indentation as the if itself ).

**Indentation and suit of a python code:**

- **Indentation** is Pythons method for associating or **grouping statements**.

- Indentation gets used in many Python statements, including the if statement.

- For if, the indentation indicates that the **indented statements** following the if are associated with that if.

- This group is called a **compound statement**, where the if keyword is part of the header and the set of indented statements is called the **suite**.

- A compound statement is considered to be one **logical statement**.

```python
# if statement
# ask the user to enter any number (integer).
# write a program to find the number is positive or negative or zero
n=int(input("Enter any number:"))
# there are three conditions, we need 3 if blocks to implement it
# positive number
if n>0:
  print("n is positive number")
# negative number
if n<0:
  print("n is negative number")
# zero
if n==0:
  print("n is neither positive nor negative ( it is zero)")
```
